```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Прочитайте файл 2017_jun_final.csv за допомогою методу read_csv.
df = pd.read_csv("2017_jun_final - 2017_jun_final.csv")
```

```python
# Прочитайте отриману таблицю, використовуючи метод head.
print(df.head())
```

```python
# Визначте розмір таблиці за допомогою методу shape.
print(df.shape)
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes.
print(df.dtypes)
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum).
print(df.isnull().sum() / len(df))
```

```python
# Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування".
df = df.drop(['Спеціалізація', 'Університет', 'Предметна область', 'Валюта', 'cls'], axis=1)
```

```python
# Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся,
# що залишився тільки стовпець "Мова.програмування".
print(df.isnull().sum() / len(df))
```

```python
# Видаліть усі рядки у вихідній таблиці за допомогою методу dropna.
df = df.dropna()
```

```python
# Визначте новий розмір таблиці за допомогою методу shape.
print(df.shape)
```

```python
# Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами,
# які вказали мову програмування Python.
python_data = df[df['Мова програмування'] == 'Python']
print(python_data)
```

```python
# Використовуючи метод groupby, виконайте групування за стовпчиком "Посада".
grouped_data = python_data.groupby('Посада')
print(grouped_data['Посада'].head())
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте 
# агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення 
# у стовпчику "Зарплата.в.місяць".
new_df = grouped_data.agg({'Зарплата на місяць': ['min', 'max']})
print(new_df)
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць.
# Використовуйте її для методу apply та створіть новий стовпчик "avg".

def fill_avg_salary(group):
    avg = group['Зарплата на місяць'].mean()
    return round(avg, 1)

new_df['avg'] = grouped_data.apply(fill_avg_salary)
print(new_df)
```

```python
# Створіть описову статистику за допомогою методу describe для нового стовпчика.
print(new_df['avg'].describe())
```

```python
# Збережіть отриману таблицю в CSV файл.
new_df['Посада'] = new_df.index
new_df.to_csv('NEW_2017_jun_final.csv', index=False)
```

```python
languages = df['Мова програмування'].value_counts(normalize=True)

other_languages = languages[languages < 0.02].sum()
languages = languages[languages >= 0.02]
languages['Інші'] = other_languages

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(languages.values, labels=languages.index, autopct='%1.1f%%', shadow=True, startangle=90)
ax.axis('equal')
plt.title('Популярність мов програмування')
plt.show()
```

```python
grouped = df.groupby('Мова програмування')['Зарплата на місяць'].mean().reset_index()
fig, ax = plt.subplots(figsize=(10, 8))

ax.bar(grouped['Мова програмування'], grouped['Зарплата на місяць'], color='Maroon')

ax.set_xlabel('Мова програмування', fontsize=15, color='red')
ax.set_xticklabels(grouped['Мова програмування'], rotation=60)

ax.set_ylabel('Середня зарплата на місяць', fontsize=15, color='red')

ax.set_title('Середні заробітні плати спеціалістів різних мов програмування', fontsize=15)
plt.show()
```
