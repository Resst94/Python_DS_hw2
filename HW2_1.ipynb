```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import requests
from io import StringIO

%matplotlib inline
```

```python
url = 'https://uk.wikipedia.org/wiki/Населення_України'
r = requests.get(url)
website = r.text.replace(',', '.')
data = pd.read_html(StringIO(website), match='Коефіцієнт народжуваності в регіонах України', encoding="UTF-8")
data = data[0]
print(data)
```

```python
# Вивести перші рядки таблиці за допомогою методу head.

data.head()
```

```python
# Визначте кількість рядків та стовпців у датафреймі (атрибут shape).

data.shape
```

```python
# Замініть у таблиці значення "—" на значення NaN.

data.replace("—", pd.NA, inplace=True)
print(data)
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes.

data.dtypes
```

```python
# Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—".

data['2014'] = pd.to_numeric(data['2014'])
data['2019'] = pd.to_numeric(data['2019'])

data.dtypes
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum).

data.isnull().sum()/data.shape[0]
```

```python
# Видаліть з таблиці дані по всій країні, останній рядок таблиці.

data = data.drop(data.index[-1])
data
```

```python
# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna).

columns_to_fill = ['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']

for column in columns_to_fill:
    data[column] = data[column].fillna(data[column].mean()).round(2)

data
```

```python
# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні.

data[data['2019'] > data['2019'].mean()][['Регіон', '2019']]
```

```python
# У якому регіоні була найвища народжуваність у 2014 році?

data[['Регіон', '2014']].max()
```

```python
# Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році.

data['2019'].plot.bar()
plt.show()
```

```python
# Додайте від 3 до 5 графіків функцій різного типу на ваш вибір. 
# Задайте графікам оформлення, щоб кожен графік у своїй домашній роботі чимось відрізнявся і не був схожим на інші.
# Можна використовувати як matplotlib, так і seaborn.
```

```python
data[['Регіон', '2012']].plot.bar(x='Регіон', y='2012', figsize=(12, 6), color='midnightblue')
plt.title('Рівень народжуваності по регіонах у 2012 році')
plt.xlabel('Регіон')
plt.ylabel('Рівень народжуваності')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
khmeln_region = data[data['Регіон'] == 'Хмельницька']
plt.figure(figsize=(10, 6))
plt.plot(data.columns[1:], khmeln_region.iloc[0, 1:], marker='^', color='red', linestyle='-.')
plt.title('Динаміка народжуваності у Хмельницькій області')
plt.xlabel('Рік')
plt.ylabel('Народжуваність')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np

plt.figure(figsize=(8, 8))
colors = plt.cm.tab20c(np.arange(len(data)))
plt.pie(data['2014'], labels=data['Регіон'], colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Кругова гістограма рівня народжуваності по регіонах у 2014 році')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for index, row in data.iterrows():
    plt.plot(data.columns[1:], row[1:], label=row['Регіон'])

plt.title('Порівняння народжуваності в регіонах протягом років')
plt.xlabel('Роки')
plt.ylabel('Коефіціент народжуваності')
plt.legend()
plt.show()
```
